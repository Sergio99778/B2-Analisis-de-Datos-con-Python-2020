```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis
```

```python
diabetes = pd.read_csv('../../Datasets/diabetes-clean.csv')
len(diabetes)
```

```python
def generate_data(column_data):
    skewness_value = skew(column_data)
    kurtosis_value = kurtosis(column_data)
    print(f'Skewness: {skewness_value}')
    print(f'Kurtosis: {kurtosis_value}')

def generate_histogram(df, column,bins=10 ):
    df.hist(column=column, bins=bins)
    plt.title(f'Histograma de {column}')
    plt.xlabel(column)
    plt.ylabel('Frecuencia')
    plt.show()
    generate_data(df[column])

for column in diabetes.columns:
    generate_histogram(diabetes, column,50)
```
